## 1. Exploring Series

```python
import pandas as pd
import numpy as np
```

```python
# Read the csv file by using nmumpy 
houses = np.genfromtxt('Houses.csv', delimiter=',', dtype=int, encoding='utf-8')
print(houses)
```

## 2. Exploring Pandas

```python
# Create Dataframe from series
group1 = pd.Series([11, 21, 31, 41, 51], index=['a', 'b', 'c', 'd', 'e'])
group2 = pd.Series([60, 70, 80, 90, 10], index=['alix', 'baa', 'c', 'd', 'edil'])
df_group = pd.DataFrame({'group1': group1, 'group2': group2})
df_group
```

```python
# Read the csv file by using pandas
df_google_apps = pd.read_csv('googleplaystore.csv')
df_google_apps

# Write a csv file
df_google_apps.to_csv('googleplaystore_copy.csv', index=False)

df_google_apps_index = df_google_apps.set_index('App')

df_google_apps_index
```

```python
# loc et iloc 
test = df_google_apps_index.loc["Coloring book moana", "Size"]
print(test)
```

```python
#ajouter une colonne 
category = df_google_apps_index["Category"]
df_google_apps_index["Category_bis"] = category
df_google_apps_index
```

```python
# Remove a column
df_google_apps_index.drop(["Category_bis","Android Ver"], axis=1)
```

### Remove null values in dataframe

```python
# count null in dataframe df_google_apps
df_google_apps.isnull().sum()

# dropo all rows with null values
df_google_apps_no_NaN = df_google_apps.dropna()
df_google_apps_no_NaN.isnull().sum()
```

### Sort dataframe 

```python
df_google_apps_no_NaN.sort_values(by='Rating',ascending=False)
```

```python
# Check type of all dataframe column
df_google_apps_no_NaN.dtypes

df_google_apps_no_NaN.describe(include="all")
```

```python
# Replace values in a column
df_google_apps_no_NaN['Installs_bis'] = df_google_apps_no_NaN['Installs'].str.replace('+','').str.replace(',','').astype(int)
df_google_apps_no_NaN
df_google_apps_no_NaN['Reviews'] = df_google_apps_no_NaN['Reviews'].astype(int)
df_google_apps_no_NaN.describe()
```
