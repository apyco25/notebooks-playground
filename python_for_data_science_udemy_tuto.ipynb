### <a id='toc1_2_1_'></a>[Remove null values in dataframe](#toc0_)

```python
# count null in dataframe df_google_apps
df_google_apps.isnull().sum()

# dropo all rows with null values
df_google_apps_no_NaN = df_google_apps.dropna()
df_google_apps_no_NaN.isnull().sum()
```

## <a id='toc1_3_'></a>[Sort dataframe](#toc0_)

```python
df_google_apps_no_NaN.sort_values(by='Rating',ascending=False)
```

## <a id='toc1_4_'></a>[Conversion data in dataframe](#toc0_)

```python
# Check type of all dataframe column
df_google_apps_no_NaN.dtypes

df_google_apps_no_NaN.describe(include="all")
```

```python
# Replace values in a column
df_google_apps_no_NaN['Installs_bis'] = df_google_apps_no_NaN['Installs'].str.replace('+','').str.replace(',','').astype(int)
df_google_apps_no_NaN
df_google_apps_no_NaN['Reviews'] = df_google_apps_no_NaN['Reviews'].astype(int)
df_google_apps_no_NaN
```

## <a id='toc1_5_'></a>[Filter in dataframe](#toc0_)

```python
# Get only df with the rating is higher than 4.5
df_google_apps_no_NaN.loc[df_google_apps_no_NaN["Rating"] > 4.5,:]
#df_google_apps_no_NaN_supp = df_google_apps_no_NaN[df_google_apps_no_NaN.loc["Rating"] > 4.5:]
```

```python
# filter applications in Business category
df_google_apps_no_NaN.loc[df_google_apps_no_NaN['Category']=='BUSINESS',:]
```

```python
# filter applications in Business category who has more than 4.5 rating
df_google_apps_no_NaN.loc[(df_google_apps_no_NaN['Category'] == 'BUSINESS') & (df_google_apps_no_NaN["Rating"] > 4.5), :]
```

## <a id='toc1_6_'></a>[Group a column in dataframe](#toc0_)

```python
# group by Price
df_google_apps_no_NaN['Price bis']= df_google_apps_no_NaN['Price'].str.replace('$','').astype(float)
df_google_apps_no_NaN.groupby(['Category','Genres']).describe()

```
